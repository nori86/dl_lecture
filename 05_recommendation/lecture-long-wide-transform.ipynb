```
import pandas as pd
```

```
d = {
    'Date':['2018-01-21','2018-01-21','2018-01-22','2018-01-22',
            '2018-01-23','2018-01-23'],
     'Name':['Smith','Mark','Joe','Mark','Smith','Joe'],
     'Sales':[200,100,300,100,250,150] }
```

```
# データフレームへ変換
df = pd.DataFrame(d, columns=['Date','Name','Sales'])
df
```

```
# long型からwide型へ変換
df2 = df.pivot(index='Name', columns='Date', values='Sales')
df2
```

```
# 欠損値の補完
df2.fillna(value=0, inplace=True)
df2
```

```
# インデックスをNameという行名にする(列を作成)
df2.reset_index(inplace=True)
df2
```

```
# pivot_tableの存在
df3 = pd.pivot_table(df, 
                     values='Sales', 
                     index='Name', 
                     columns='Date',
                     aggfunc='mean', 
                     fill_value=0)
df3
```

```
# インデックスをNameという行名にする(列を作成)
df3.reset_index(inplace=True)
df3
```

```
# データセット2の作成
d2 = {
'Date':['2018-01-21','2018-01-21','2018-01-21','2018-01-22','2018-01-22','2018-01-22',
'2018-01-23','2018-01-23'],
'Name':['Smith','Smith','Mark','Joe','Mark','Mark','Smith','Joe'],
'Sales':[200,400,100,300,100,300,250,150] }
```

```
# データフレームへ変換
df4 = pd.DataFrame(d2, columns=['Date','Name','Sales'])
df4
```

```
# 重複がある場合、pivotの前にgroupbyする
df5 = df4.groupby(by=['Name','Date'], as_index=False, sort=False)
df5 = df5.sum()
df5
```

```
# pivotでのlong→wide変換
df6 = df5.pivot(index='Name', columns='Date', values='Sales')
df6
```

```
# 欠損値の補完
df6.fillna(value=0, inplace=True)
df6
```

```
# インデックスをNameという行名にする(列を作成)
df6.reset_index(inplace=True)
df6
```

```
# pivot_tableでのlong→wide変換
df7 = pd.pivot_table(df4, values='Sales', index='Name', columns='Date', aggfunc='sum', fill_value=0)
df7
```

```
# インデックスをNameという行名にする(列を作成)
df7.reset_index(inplace=True)
df7
```
