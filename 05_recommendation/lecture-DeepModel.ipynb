```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(200,activation = 'relu',input_shape = (trainX.shape[1],)))
model.add(Dropout(0.5))
model.add(Dense(200,activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(trainy.shape[1],activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'rmsprop',
              loss = 'binary_crossentropy', 
              metrics = ['accuracy'])
```

```python
model.fit(trainX,trainy,
          batch_size = 64, 
          epochs = 20,
          verbose = 1,
          validation_data = (valX,valy))
```

```python
#モデルの構造とパラメータ数
model.summary()
```

```python
#各顧客へのおすすめカテゴリを格納するndarrayの入れ物を作成
#各顧客にn個商品を薦める
n=20
#今回予測する顧客数(gyou)
gyou =  a.shape[0]
#結果を格納する箱
result = np.zeros((gyou,n))
#頭出し
result[0:3,0:10]
```
