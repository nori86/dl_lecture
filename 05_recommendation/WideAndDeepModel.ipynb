```python
trainy = train[:,0:w]
trainX = train[:,w:3*w]
trainX2 = train[:,3*w:]
valy = val[:,0:w]
valX = val[:,w:3*w]
valX2 = val[:,3*w:]
```

```python
print(trainy.shape)
print(trainX.shape)
print(trainX2.shape)
print(valy.shape)
print(valX.shape)
print(valX2.shape)
```

```python
import keras
```

```python
product_input = keras.layers.Input(shape=[2*w], name='Product')
```

```python
dense_1 = keras.layers.Dense(200,name='FullyConnected-1')(product_input)
dense_batch1 = keras.layers.BatchNormalization(name='Batch')(dense_1)
dropout_1 = keras.layers.Dropout(0.2,name='Dropout-1')(dense_batch1)
dense_2 = keras.layers.Dense(100,name='FullyConnected-2')(dropout_1)
dense_batch_2 = keras.layers.BatchNormalization(name='Batch-2')(dense_2)
dropout_2 = keras.layers.Dropout(0.2,name='Dropout-2')(dense_batch_2)
```

```python
age_input = keras.layers.Input(shape=[10], name='Age')
```

```python
concat = keras.layers.concatenate([dropout_2, age_input],name='Concat')
```

```python
pred = keras.layers.Dense(trainy.shape[1], activation='sigmoid', name='FinalLayer')(concat)
```

```python
#Input_and_Output
model = keras.Model([product_input, age_input], pred)
```

```python
#Optimizer
#opt = keras.optimizers.Adam(lr =0.01)
```

```python
#Model_Compiling
model.compile(optimizer='adam',loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit([trainX, trainX2], trainy, epochs=20, verbose=1, batch_size = 64, validation_data = ([valX,valX2],valy))
```
