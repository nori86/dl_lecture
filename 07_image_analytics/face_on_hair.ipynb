```
#coding: utf-8
import cv2
from PIL import Image

cascade_file = "haarcascade_frontalface_alt.xml"
image_file = "woman_001.png"

img = cv2.imread(image_file)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cascade = cv2.CascadeClassifier(cascade_file)
face_list = cascade.detectMultiScale(img_gray, minSize=(150, 150))

if len(face_list) == 0:
  print("Fail recognise")
  quit()

for (x, y, w, h) in face_list:
  print("顔の座標 =", x, y, w, h)
  color = (0, 0, 225)
  pen_w = 8
  cv2.rectangle(img, (x, y), (x+w, y+h), color, thickness = pen_w)

cv2.imwrite("woman_001_rectangle.png", img)
```

```
# わかりやすいように、face_listの値をまとめておく
x = face_list[0][0]  # X座標
y = face_list[0][1]  # Y座標
w = face_list[0][2]  # 横幅
h = face_list[0][3]  # 縦幅
```

```
# PILで画像を開く
im1 = Image.open('woman_001.png')
im2 = Image.open('woman-hair001.png')
 
# 顔に合ったサイズにニンニク画像をリサイズする
im22 = im2.resize((int(w*1.3), int(h*1.3)))
#im22.save('woman-hair001_1.png')
 
# 原本はそのままにしておくため、コピーを加工する
back_im = im1.copy()
 
# 背景透過のためのsplit()
back_im.paste(im22, (x-int(w/4), y-int(h/2)),im22.split()[3])
 
# 保存する
back_im.save('woman_001_hair_001.png')
```
