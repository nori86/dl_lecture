```
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib inline
```

```
mu = 60
sd = 10
```

```
# 確率密度の描画
xx = np.linspace(30, 90, 301)
yy = st.norm.pdf(xx, mu, sd)
plt.ylim(0, .045)
plt.plot(xx, yy)
```

```
# 生成された乱数のヒストグラム描画
x = np.random.normal(mu, sd, size=200)
plt.hist(x, bins=20)
plt.show()
```
